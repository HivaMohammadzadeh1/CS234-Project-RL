```python
import os
import math
import hashlib
import requests

from IPython.display import Markdown, display
import numpy as np
from PIL import Image
import decord
from decord import VideoReader, cpu


def download_video(url, dest_path):
    response = requests.get(url, stream=True)
    with open(dest_path, 'wb') as f:
        for chunk in response.iter_content(chunk_size=8096):
            f.write(chunk)
    print(f"Video downloaded to {dest_path}")


def get_video_frames(video_path, num_frames=128, cache_dir='.cache'):
    os.makedirs(cache_dir, exist_ok=True)

    video_hash = hashlib.md5(video_path.encode('utf-8')).hexdigest()
    if video_path.startswith('http://') or video_path.startswith('https://'):
        video_file_path = os.path.join(cache_dir, f'{video_hash}.mp4')
        if not os.path.exists(video_file_path):
            download_video(video_path, video_file_path)
    else:
        video_file_path = video_path

    frames_cache_file = os.path.join(cache_dir, f'{video_hash}_{num_frames}_frames.npy')
    timestamps_cache_file = os.path.join(cache_dir, f'{video_hash}_{num_frames}_timestamps.npy')

    if os.path.exists(frames_cache_file) and os.path.exists(timestamps_cache_file):
        frames = np.load(frames_cache_file)
        timestamps = np.load(timestamps_cache_file)
        return video_file_path, frames, timestamps

    vr = VideoReader(video_file_path, ctx=cpu(0))
    total_frames = len(vr)

    indices = np.linspace(0, total_frames - 1, num=num_frames, dtype=int)
    frames = vr.get_batch(indices).asnumpy()
    timestamps = np.array([vr.get_frame_timestamp(idx) for idx in indices])

    np.save(frames_cache_file, frames)
    np.save(timestamps_cache_file, timestamps)

    return video_file_path, frames, timestamps


def create_image_grid(images, num_columns=8):
    pil_images = [Image.fromarray(image) for image in images]
    num_rows = math.ceil(len(images) / num_columns)

    img_width, img_height = pil_images[0].size
    grid_width = num_columns * img_width
    grid_height = num_rows * img_height
    grid_image = Image.new('RGB', (grid_width, grid_height))

    for idx, image in enumerate(pil_images):
        row_idx = idx // num_columns
        col_idx = idx % num_columns
        position = (col_idx * img_width, row_idx * img_height)
        grid_image.paste(image, position)

    return grid_image
```

Inference function

```python
from qwen_vl_utils import process_vision_info
def inference(video, prompt, max_new_tokens=2048, total_pixels=20480 * 32 * 32, min_pixels=64 * 32 * 32, max_frames= 2048, sample_fps = 2):
    """
    Perform multimodal inference on input video and text prompt to generate model response.

    Args:
        video (str or list/tuple): Video input, supports two formats:
            - str: Path or URL to a video file. The function will automatically read and sample frames.
            - list/tuple: Pre-sampled list of video frames (PIL.Image or url).
              In this case, `sample_fps` indicates the frame rate at which these frames were sampled from the original video.
        prompt (str): User text prompt to guide the model's generation.
        max_new_tokens (int, optional): Maximum number of tokens to generate. Default is 2048.
        total_pixels (int, optional): Maximum total pixels for video frame resizing (upper bound). Default is 20480*32*32.
        min_pixels (int, optional): Minimum total pixels for video frame resizing (lower bound). Default is 16*32*32.
        sample_fps (int, optional): ONLY effective when `video` is a list/tuple of frames!
            Specifies the original sampling frame rate (FPS) from which the frame list was extracted.
            Used for temporal alignment or normalization in the model. Default is 2.

    Returns:
        str: Generated text response from the model.

    Notes:
        - When `video` is a string (path/URL), `sample_fps` is ignored and will be overridden by the video reader backend.
        - When `video` is a frame list, `sample_fps` informs the model of the original sampling rate to help understand temporal density.
    """

    messages = [
        {"role": "user", "content": [
                {"video": video,
                "total_pixels": total_pixels,
                "min_pixels": min_pixels,
                "max_frames": max_frames,
                'sample_fps':sample_fps},
                {"type": "text", "text": prompt},
            ]
        },
    ]
    text = processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    image_inputs, video_inputs, video_kwargs = process_vision_info([messages], return_video_kwargs=True,
                                                                   image_patch_size= 16,
                                                                   return_video_metadata=True)
    if video_inputs is not None:
        video_inputs, video_metadatas = zip(*video_inputs)
        video_inputs, video_metadatas = list(video_inputs), list(video_metadatas)
    else:
        video_metadatas = None
    inputs = processor(text=[text], images=image_inputs, videos=video_inputs, video_metadata=video_metadatas, **video_kwargs, do_resize=False, return_tensors="pt")
    inputs = inputs.to('cuda')

    output_ids = model.generate(**inputs, max_new_tokens=max_new_tokens)
    generated_ids = [output_ids[len(input_ids):] for input_ids, output_ids in zip(inputs.input_ids, output_ids)]
    output_text = processor.batch_decode(generated_ids, skip_special_tokens=True, clean_up_tokenization_spaces=True)
    return output_text[0]
```

# Video Model Setup


```python
import torch.nn as nn
class BradleyTerryRewardModel(nn.Module):
    def __init__(self, base_lm):
        super().__init__()
        self.lm = base_lm
        self.config = base_lm.config
        self.head = nn.Sequential(
            nn.Linear(self.lm.config.text_config.hidden_size, 1),
            # nn.Sigmoid()
        )
        nn.init.normal_(self.head[0].weight, std=0.01)
        nn.init.zeros_(self.head[0].bias)

    def _sequence_rep(self, hidden, attention_mask):
        lengths = attention_mask.sum(dim=1) - 1
        batch_idx = torch.arange(hidden.size(0), device=hidden.device)
        return hidden[batch_idx, lengths]

    def forward(self, input_ids, attention_mask, pixel_values_videos, video_grid_thw, **kwargs):
        # attention_mask = inputs['attention_mask']
        # inputs.pop('output_hidden_states', None)
        outputs = self.lm(
            input_ids=input_ids,
        attention_mask=attention_mask,
        pixel_values_videos=pixel_values_videos,
        video_grid_thw=video_grid_thw,
            output_hidden_states=True,
            return_dict=True,
        )
        hidden = outputs.hidden_states[-1]
        seq_repr = self._sequence_rep(hidden, attention_mask)
        return self.head(seq_repr).squeeze(-1)
```

```python

from qwen_vl_utils import process_vision_info

def vl_inference_batch(videos, prompts, max_new_tokens=2048, total_pixels=5120 * 32 * 32, min_pixels=32 * 32 * 32, max_frames= 64, sample_fps = 32):
    #prompt_template = "Evaluate the following video based on physical accuracy, temporal consistency, realism of motion, and whether it accurately captures gravity, for the prompt: {prompt}. Giving a real-number score from 0 (worse) to 1 (best)." # Output your prediction in a JSON object, for example: {{score: 0.50}}"
    prompt_template = "Evaluate the following video based on physical accuracy, temporal consistency, realism of motion, and whether it accurately captures gravity. Giving a real-number score from 0 (worse) to 1 (best)." # Output your prediction in a JSON object, for example: {{score: 0.50}}"

    all_messages = [
        [{"role": "user", "content": [
                {"video": video,
                "total_pixels": total_pixels,
                "min_pixels": min_pixels,
                "max_frames": max_frames,
                'sample_fps':sample_fps},
                {"type": "text", "text": prompt_template},
            ]
        },] for video, prompt in zip(videos,prompts)
    ]
    text = [processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=True) for messages in all_messages]
    image_inputs, video_inputs, video_kwargs = process_vision_info(all_messages, return_video_kwargs=True,
                                                                   image_patch_size= 16,
                                                                   return_video_metadata=True) # batch should work
    if video_inputs is not None:
        video_inputs, video_metadatas = zip(*video_inputs)
        video_inputs, video_metadatas = list(video_inputs), list(video_metadatas)
    else:
        video_metadatas = None
    inputs = processor(text=text, images=image_inputs, videos=video_inputs, video_metadata=video_metadatas, **video_kwargs, do_resize=False, padding= True,return_tensors="pt")
    del video_inputs, image_inputs, text
    inputs = inputs.to('cuda')
    return inputs

def vl_prompt_batch(videos, prompts, max_new_tokens=2048, total_pixels=5120 * 32 * 32, min_pixels=32 * 32 * 32, max_frames= 64, sample_fps = 32):
    prompt_template = "Evaluate the following video based on physical accuracy, temporal consistency, realism of motion, and whether it accurately captures gravity, for the prompt: {prompt}. Giving a real-number score from 0 (worse) to 1 (best). Use the **full range** of scores (0.0, 0.01, 0.02, ..., 0.99, 1.0) as appropriate, and critically evaluate every frame strictly. Output your prediction in a JSON object, for example: {{score: 0.16}}"
    all_messages = [
        [{"role": "user", "content": [
                {"video": video,
                "total_pixels": total_pixels,
                "min_pixels": min_pixels,
                "max_frames": max_frames,
                'sample_fps':sample_fps},
                {"type": "text", "text": prompt_template.format(prompt = prompt)},
            ]
        },] for video, prompt in zip(videos,prompts)
    ]
    text = [processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=True) for messages in all_messages]
    image_inputs, video_inputs, video_kwargs = process_vision_info(all_messages, return_video_kwargs=True,
                                                                   image_patch_size= 16,
                                                                   return_video_metadata=True) # batch should work
    if video_inputs is not None:
        video_inputs, video_metadatas = zip(*video_inputs)
        video_inputs, video_metadatas = list(video_inputs), list(video_metadatas)
    else:
        video_metadatas = None
    inputs = processor(text=text, images=image_inputs, videos=video_inputs, video_metadata=video_metadatas, **video_kwargs, do_resize=False, padding= True,return_tensors="pt")
    del video_inputs, image_inputs, text
    inputs = inputs.to('cuda')
    return inputs


def vl_prompt(video, prompt, max_new_tokens=2048, total_pixels=20480 * 32 * 32, min_pixels=64 * 32 * 32, max_frames= 2048, sample_fps = 32):
    """
    Perform multimodal inference on input video and text prompt to generate model response.

    Args:
        video (str or list/tuple): Video input, supports two formats:
            - str: Path or URL to a video file. The function will automatically read and sample frames.
            - list/tuple: Pre-sampled list of video frames (PIL.Image or url).
              In this case, `sample_fps` indicates the frame rate at which these frames were sampled from the original video.
        prompt (str): User text prompt to guide the model's generation.
        max_new_tokens (int, optional): Maximum number of tokens to generate. Default is 2048.
        total_pixels (int, optional): Maximum total pixels for video frame resizing (upper bound). Default is 20480*32*32.
        min_pixels (int, optional): Minimum total pixels for video frame resizing (lower bound). Default is 16*32*32.
        sample_fps (int, optional): ONLY effective when `video` is a list/tuple of frames!
            Specifies the original sampling frame rate (FPS) from which the frame list was extracted.
            Used for temporal alignment or normalization in the model. Default is 2.

    Returns:
        str: Generated text response from the model.

    Notes:
        - When `video` is a string (path/URL), `sample_fps` is ignored and will be overridden by the video reader backend.
        - When `video` is a frame list, `sample_fps` informs the model of the original sampling rate to help understand temporal density.
    """
    prompt_template = f"Evaluate the following video based on physical accuracy, temporal consistency, and realism of motion for the prompt: {prompt}. Assign a real-number score from 0 to 1."
    messages = [
        {"role": "user", "content": [
                {"video": video,
                "total_pixels": total_pixels,
                "min_pixels": min_pixels,
                "max_frames": max_frames,
                'sample_fps':sample_fps},
                {"type": "text", "text": prompt_template},
            ]
        },
    ]
    text = processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=True, padding="max_length",     # pad to max_length
        truncation=True,
        max_length=2048)
    image_inputs, video_inputs, video_kwargs = process_vision_info([messages], return_video_kwargs=True,
                                                                   image_patch_size= 16,
                                                                   return_video_metadata=True)
    if video_inputs is not None:
        video_inputs, video_metadatas = zip(*video_inputs)
        video_inputs, video_metadatas = list(video_inputs), list(video_metadatas)
    else:
        video_metadatas = None
    inputs = processor(text=[text], images=image_inputs, videos=video_inputs, video_metadata=video_metadatas, **video_kwargs, do_resize=False, padding="max_length", truncation=True, max_length = 2048,return_tensors="pt")
    inputs = inputs.to('cuda')
    return inputs
```

```python
import torch
import numpy as np
import gc
import os
from diffusers import WanPipeline, AutoencoderKLWan
from diffusers.utils import export_to_video

# ── Config ────────────────────────────────────────────────────────────────────
dtype = torch.bfloat16
device = "cuda"
model_id = "Wan-AI/Wan2.2-TI2V-5B-Diffusers"

HEIGHT = 704
WIDTH = 1280
NUM_FRAMES = 81
NUM_INFERENCE_STEPS = 20
GUIDANCE_SCALE = 3.5
FPS = 16
OUTPUT_DIR = "outputs"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ── Model loading ─────────────────────────────────────────────────────────────
vae = AutoencoderKLWan.from_pretrained(model_id, subfolder="vae", torch_dtype=torch.float32)
pipe = WanPipeline.from_pretrained(model_id, vae=vae, torch_dtype=dtype)
pipe.to(device)


# ── Inference loop ────────────────────────────────────────────────────────────
# prompts = [
#     "A tennis ball falling from 2 meters onto a hard concrete floor and bouncing several times. Static camera, side view.",
#     # "A rubber ball dropped onto a wooden floor, slow motion, close-up.",
#     # # add more prompts here...
# ]

results = []

for i, batch in enumerate(dataloader):
    prompts = batch['prompt']          # list of strings (batch_size,)
    video_paths = []

    # ── Step 1: Generate videos for the batch ─────────────────────────────
    for j, prompt in enumerate(prompts):
        frames = pipe(
            prompt=prompt,
            height=HEIGHT,
            width=WIDTH,
            num_frames=NUM_FRAMES,
            guidance_scale=GUIDANCE_SCALE,
            num_inference_steps=NUM_INFERENCE_STEPS,
        ).frames[0]

        video_path = os.path.join(OUTPUT_DIR, f"video_{i:04d}_{j:02d}.mp4")
        export_to_video(frames, video_path, fps=FPS)
        video_paths.append(video_path)
        del frames

    # ── Step 2: Score the whole batch at once ──────────────────────────────
    video_inputs = vl_prompt_batch(
        videos=video_paths,
        prompts=list(prompts),
        sample_fps=64,
    )

    with torch.no_grad():
        rewards = lora_rm(**video_inputs)   # shape: (batch_size,)

    print(f"Batch {i} rewards: {rewards}")

    # ── Cleanup ────────────────────────────────────────────────────────────
    del video_inputs, rewards
    torch.cuda.empty_cache()
    gc.collect()

# ── Summary ───────────────────────────────────────────────────────────────────
print("\n── Results ──────────────────────────────────────────────────────────────")
for r in sorted(results, key=lambda x: x["reward"], reverse=True):
    print(f"  [{r['reward']:.4f}]  {r['prompt'][:80]}")
    print(f"           {r['video_path']}")
```

```python
import torch
import torch.nn.functional as F
import gc
from copy import deepcopy
from diffusers import UniPCMultistepScheduler
from diffusers.utils import export_to_video


# ─────────────────────────────────────────────────────────────────────────────
# Helper: run one full denoising trajectory and record log probs at every step
# ─────────────────────────────────────────────────────────────────────────────
def sample_with_log_probs(pipe, prompt, height, width, num_frames, cfg):
    """
    Returns:
        frames      : decoded video frames (for reward model)
        log_probs   : tensor [T] — sum of log probs across denoising steps
        latents_traj: list of latents at each step (for re-evaluation)
        noise_preds : list of noise predictions at each step
        timesteps   : scheduler timesteps used
    """
    scheduler = pipe.scheduler
    scheduler.set_timesteps(cfg["num_inference_steps"])
    timesteps = scheduler.timesteps

    # Encode prompt
    with torch.no_grad():
        prompt_embeds, neg_embeds = pipe.encode_prompt(
            prompt, do_classifier_free_guidance=True, negative_prompt=""
        )
        prompt_embeds = prompt_embeds.to(device)
        neg_embeds = neg_embeds.to(device)

    # Initial noise
    latent_shape = (
        1,
        pipe.transformer.config.in_channels,
        (num_frames - 1 ) // pipe.vae.config.scale_factor_temporal + 1,
        height // pipe.vae.config.scale_factor_spatial,
        width  // pipe.vae.config.scale_factor_spatial,
    )
    latents = torch.randn(latent_shape, device=device, dtype=dtype)
    latents = latents * scheduler.init_noise_sigma

    latents_traj = []
    # noise_preds  = []
    log_probs    = []

    total_steps = len(timesteps)  # add this before the loop
    K = 5  # add this before the loop

    for i, t in enumerate(timesteps):
        # print(t)
        # exit(0)
        latents_traj.append(latents.detach().cpu().clone())

        # CFG: concatenate unconditional + conditional
        latent_input = torch.cat([latents] * 2)
        latent_input = scheduler.scale_model_input(latent_input, t)

        # Forward through transformer (with grad for training, without for sampling)
        with torch.no_grad():
            noise_pred = pipe.transformer(
                hidden_states=latent_input.to(device),
                timestep=t.unsqueeze(0).to(device),
                encoder_hidden_states=torch.cat([neg_embeds, prompt_embeds]).to(device),
            ).sample

        del latent_input

        # CFG guidance
        noise_uncond, noise_cond = noise_pred.chunk(2)
        noise_pred = noise_uncond + 3.5 * (noise_cond - noise_uncond)
        # noise_preds.append(noise_pred.detach().clone())
        del noise_uncond, noise_cond

        # Compute log prob of this denoising step
        # p(x_{t-1} | x_t) is Gaussian with mean = scheduler_step_mean, std = sigma_t
        # prev_latents = scheduler.step(noise_pred, t, latents, eta=cfg["eta"]).prev_sample
        prev_latents = scheduler.step(noise_pred, t, latents).prev_sample
        # del noise_pred
        # Log prob: -0.5 * ||x_{t-1} - mu||^2 / sigma^2
        # alpha_prod    = scheduler.alphas_cumprod[t]
        # beta_prod     = 1 - alpha_prod
        # sigma_t       = cfg["eta"] * (beta_prod ** 0.5)
        # sigma_t       = max(sigma_t, 1e-6)  # numerical stability

        # log_prob = -0.5 * ((prev_latents - latents) ** 2).sum() / (sigma_t ** 2)
        # log_probs.append(log_prob)

        # sigma_t = 1.0
        # log_prob = -0.5 * ((noise_pred) ** 2).sum() / (sigma_t ** 2)

        # try this to avoid numerical explosion
        # only store last 5 steps
        if i >= total_steps - K:
            n_elements = noise_pred.numel()
            sigma_t = 1.0
            log_prob = -0.5 * (noise_pred ** 2).sum() / (sigma_t ** 2 * n_elements)

            log_probs.append(log_prob)

        latents = prev_latents
        del prev_latents, noise_pred
        torch.cuda.empty_cache()

    # Decode to video
    with torch.no_grad():
        latents_mean = torch.tensor(pipe.vae.config.latents_mean).to(device, dtype=dtype).view(1, -1, 1, 1, 1)
        latents_std = torch.tensor(pipe.vae.config.latents_std).to(device, dtype=dtype).view(1, -1, 1, 1, 1)

        # denormalize before decoding
        latents_decoded = latents.float() * latents_std + latents_mean
        print(f"Max: {latents_decoded.max().item()}, Min: {latents_decoded.min().item()}")
        video = pipe.vae.decode(latents_decoded).sample
        # video = pipe.vae.decode(latents.float()).sample
        video = (video / 2 + 0.5).clamp(0, 1)
        video = video.permute(0, 2, 1, 3, 4).squeeze(0)  # [T, C, H, W]
        frames = (video.cpu().float().numpy() * 255).astype("uint8")
        frames = [f.transpose(1, 2, 0) for f in frames]

    total_log_prob = torch.stack(log_probs).sum()
    return frames, total_log_prob, latents_traj, list(timesteps)


# ─────────────────────────────────────────────────────────────────────────────
# Helper: re-evaluate log probs under current policy (for importance sampling)
# ─────────────────────────────────────────────────────────────────────────────
# def evaluate_log_probs(pipe, latents_traj, noise_preds_old, timesteps, prompt_embeds, neg_embeds, cfg):
def evaluate_log_probs(pipe, latents_traj, timesteps, prompt_embeds, neg_embeds, cfg):
    """Re-run the transformer over the stored trajectory to get new log probs."""
    scheduler = pipe.scheduler
    scheduler.set_timesteps(cfg["num_inference_steps"])
    total_steps = len(timesteps)
    K = 5 # try

    log_probs = []

    for i, t in enumerate(timesteps):
        if i < total_steps - K:
          continue  # skip — don't run transformer at all
        latents    = latents_traj[i].to(device)
        latent_input = torch.cat([latents] * 2)
        # latent_input = scheduler.scale_model_input(latent_input, t)
        latent_input = scheduler.scale_model_input(latent_input, t)

        # This forward pass *does* compute gradients
        # noise_pred = pipe.transformer(
        #     hidden_states=latent_input,
        #     timestep=t.unsqueeze(0).to(device),
        #     # encoder_hidden_states=...,   # cache prompt embeds outside this fn
        #     encoder_hidden_states=torch.cat([neg_embeds, prompt_embeds]).to(device),
        # ).sample
        # if i < total_steps - K:
        #     with torch.no_grad():
        #         noise_pred = pipe.transformer(
        #             hidden_states=latent_input,
        #             timestep=t.unsqueeze(0).to(device),
        #             encoder_hidden_states=torch.cat([neg_embeds, prompt_embeds]).to(device),
        #         ).sample
        # else:
        # if i < total_steps - K:
        #     # no grad for early steps
        #     with torch.no_grad():
        #         noise_pred = pipe.transformer(
        #             hidden_states=latent_input,
        #             timestep=t.unsqueeze(0).to(device),
        #             encoder_hidden_states=torch.cat([neg_embeds, prompt_embeds]).to(device),
        #         ).sample
        #     noise_uncond, noise_cond = noise_pred.chunk(2)
        #     noise_pred = noise_uncond + 3.5 * (noise_cond - noise_uncond)
        #     del noise_uncond, noise_cond

        #     n_elements = noise_pred.numel()
        #     sigma_t = 1.0
        #     log_prob = -0.5 * (noise_pred ** 2).sum() / (sigma_t ** 2 * n_elements)
        #     # detach so no grad accumulates from early steps
        #     log_probs.append(log_prob.detach())

        # else:
        # grad for last K steps
        noise_pred = pipe.transformer(
            hidden_states=latent_input,
            timestep=t.unsqueeze(0).to(device),
            encoder_hidden_states=torch.cat([neg_embeds, prompt_embeds]).to(device),
        ).sample
        noise_uncond, noise_cond = noise_pred.chunk(2)
        noise_pred = noise_uncond + 3.5 * (noise_cond - noise_uncond)
        del noise_uncond, noise_cond

        n_elements = noise_pred.numel()
        sigma_t = 1.0
        log_prob = -0.5 * (noise_pred ** 2).sum() / (sigma_t ** 2 * n_elements)
        log_probs.append(log_prob)  # keep grad


        del latents, noise_pred
        torch.cuda.empty_cache()

    return torch.stack(log_probs).sum()

```

```python
# ── DDPO Config ───────────────────────────────────────────────────────────────
DDPO_CFG = dict(
    num_epochs          = 2, #2,
    sample_batch_size   = 1, #2, #2,       # videos generated per step
    train_batch_size    = 1, #2, #2,       # minibatch size for gradient update
    num_inner_epochs    = 2, #1,       # gradient steps per sample batch
    clip_range          = 0.2,     # PPO-style clipping (DDPO-IS variant)
    lr                  = 1.5e-4,
    kl_coeff            = 0.00,    # don't need kl since we already do clipping.
    grad_clip           = 1.0,
    num_inference_steps = 20, #15, #20, # try first
    # eta                 = 1.0,     # stochasticity in scheduler (must be >0 for log probs)
)

# ── Scheduler must be DDIM with eta > 0 for tractable log probs ───────────────
# Replace the pipe's scheduler
from diffusers import UniPCMultistepScheduler, DDIMScheduler
# pipe.scheduler = DDIMScheduler.from_config(pipe.scheduler.config)
pipe.scheduler = UniPCMultistepScheduler.from_config(pipe.scheduler.config)
pipe.scheduler.set_timesteps(DDPO_CFG["num_inference_steps"])

# ── Frozen reference model (for KL penalty) ───────────────────────────────────
# Keep a copy of the transformer *before* LoRA for KL
ref_transformer = deepcopy(pipe.transformer)
ref_transformer.requires_grad_(False)
ref_transformer.eval()

# ── Optimizer over LoRA params only ───────────────────────────────────────────
optimizer = torch.optim.AdamW(
    filter(lambda p: p.requires_grad, pipe.transformer.parameters()),
    lr=DDPO_CFG["lr"],
)
```

```python
# Testing
import itertools
def chunks(iterable, n):
    it = iter(iterable)
    while True:
        chunk = list(itertools.islice(it, n))
        if not chunk:
            break
        yield chunk

for chunk_idx, batch_chunk in enumerate(chunks(dataloader, 8)):
    print(batch_chunk)
    print(len(batch_chunk))
    print(batch_chunk[0])
```

```python
import wandb
run = wandb.init(
    entity="shdkwok-stanford",
    project="cs234",
    config=DDPO_CFG,
    # name=f"ddpo-wan2.2-{wandb.util.generate_id()}",
    # settings=wandb.Settings(start_method="thread"),  # prevents hanging on Colab
)


for epoch in range(DDPO_CFG["num_epochs"]):
    print(f"\n═══ Epoch {epoch} ═══")
    for chunk_idx, batch_chunk in enumerate(chunks(dataloader, 8)):
        print(f"\n═══ Chunk {chunk_idx} ═══")
        # swap: pipe on GPU, lora_rm on CPU
        pipe.transformer.to(device) #cbb
        pipe.transformer.enable_gradient_checkpointing()
        pipe.vae.to(device)
        pipe.text_encoder.to(device)
        lora_rm.cpu()
        torch.cuda.empty_cache()

        # ── Phase 1a: Sample ─────────────────────────────────────────────────────
        sample_buffer = []
        video_paths = []
        for idx, batch in enumerate(batch_chunk):
            prompts = batch['prompt']
            print(f"Loading batch {idx}")
            for prompt in prompts:
                for _ in range(DDPO_CFG["sample_batch_size"]):
                    try:
                        frames, log_prob_old, latents_traj, timesteps = sample_with_log_probs(
                            pipe, prompt, height=704, width=1280, num_frames=81, cfg=DDPO_CFG
                        )
                        path = f"/tmp/tmp_ddpo_{epoch}_{len(sample_buffer)}.mp4"
                        export_to_video(frames, path, fps=16)
                        del frames
                        torch.cuda.empty_cache()

                        sample_buffer.append({
                            "prompt":       prompt,
                            "log_prob_old": log_prob_old.detach(),
                            "latents_traj": latents_traj,
                            # "noise_preds":  noise_preds,
                            "timesteps":    timesteps,
                        })
                        video_paths.append(path)

                    except torch.cuda.OutOfMemoryError:
                        print(f"  OOM during sampling, skipping")
                        if 'frames' in dir(): del frames
                        torch.cuda.empty_cache(); gc.collect()
                        continue
                    except Exception as e:
                        print(f"  Error during sampling: {e}, skipping")
                        torch.cuda.empty_cache(); gc.collect()
                        continue

        if len(sample_buffer) == 0:
            print("  No samples collected, skipping epoch")
            continue

        # ── Phase 1b: Score (swap to lora_rm on GPU) ─────────────────────────────
        pipe.transformer.cpu()
        pipe.vae.cpu()
        pipe.text_encoder.cpu()

        torch.cuda.empty_cache()
        lora_rm.to(device)

        score_batch_size = 2  # tune this based on memory

        for i in range(0, len(sample_buffer), score_batch_size):
            batch_samples = sample_buffer[i:i+score_batch_size]
            batch_paths = video_paths[i:i+score_batch_size]

            try:
                with torch.no_grad():
                    batch_inputs = vl_prompt_batch(
                        videos=batch_paths,
                        prompts=[s["prompt"] for s in batch_samples],
                        sample_fps=64
                    )
                    batch_rewards = lora_rm(**batch_inputs)
                    del batch_inputs
                torch.cuda.empty_cache()

                for sample, reward in zip(batch_samples, batch_rewards):
                    sample["reward"] = reward.detach()
                    print(f"  [score {i}] reward={reward.item():.4f}")

            except torch.cuda.OutOfMemoryError:
                print(f"  OOM scoring batch {i}-{i+score_batch_size}, falling back to one by one")
                for sample, path in zip(batch_samples, batch_paths):
                    try:
                        with torch.no_grad():
                            video_inputs = vl_prompt_batch(videos=[path], prompts=[sample["prompt"]], sample_fps=64)
                            reward = lora_rm(**video_inputs).squeeze()
                            del video_inputs
                        sample["reward"] = reward.detach()
                        print(f"  [score fallback] reward={reward.item():.4f}")
                    except torch.cuda.OutOfMemoryError:
                        print(f"  OOM on single sample, assigning zero reward")
                        sample["reward"] = torch.tensor(0.0)
                        torch.cuda.empty_cache(); gc.collect()
                    finally:
                        torch.cuda.empty_cache()
            finally:
                for path in batch_paths:
                    if os.path.exists(path):
                        os.remove(path)
                torch.cuda.empty_cache()

        lora_rm.cpu()
        torch.cuda.empty_cache()

        # ── Normalize rewards ─────────────────────────────────────────────────────
        rewards = torch.stack([s["reward"] for s in sample_buffer])
        wandb.log({
            "epoch": epoch,
            "reward_mean": rewards.mean().item(),
            "reward_std": rewards.std().item(),
            "reward_min": rewards.min().item(),
            "reward_max": rewards.max().item(),
            "num_samples": len(sample_buffer),
        })
        rewards = (rewards - rewards.mean()) / (rewards.std() + 1e-12)
        for s, r in zip(sample_buffer, rewards):
            s["reward_normalized"] = r

        # rewards_raw = torch.stack([s["reward"] for s in sample_buffer])


        # ── Phase 2: PPO update (swap pipe back to GPU) ───────────────────────────
        pipe.transformer.to(device)
        pipe.text_encoder.to(device)
        pipe.transformer.enable_gradient_checkpointing()
        torch.cuda.empty_cache()
        pipe.transformer.train()

        # optimizer = torch.optim.AdamW(
        #     filter(lambda p: p.requires_grad, pipe.transformer.parameters()),
        #     lr=DDPO_CFG["lr"],
        # )

        for group in optimizer.param_groups:
            group['params'] = [p for p in pipe.transformer.parameters() if p.requires_grad]

        total_loss = 0
        update_count = 0

        for inner_epoch in range(DDPO_CFG["num_inner_epochs"]):
            for sample in sample_buffer:
                try:
                    # debug check
                    if len(sample["timesteps"]) == 0 or len(sample["latents_traj"]) == 0:
                        print(f"  Empty timesteps or latents_traj, skipping sample")
                        continue
                    optimizer.zero_grad()

                    with torch.no_grad():
                        prompt_embeds, neg_embeds = pipe.encode_prompt(
                            sample["prompt"], do_classifier_free_guidance=True, negative_prompt=""
                        )
                        prompt_embeds = prompt_embeds.to(device)
                        neg_embeds = neg_embeds.to(device)

                    log_prob_new = evaluate_log_probs(
                        pipe,
                        sample["latents_traj"],
                        sample["timesteps"],
                        prompt_embeds,
                        neg_embeds,
                        DDPO_CFG,
                    )

                    del prompt_embeds, neg_embeds
                    torch.cuda.empty_cache()

                    log_prob_diff = log_prob_new - sample["log_prob_old"]
                    print(f"  log_prob_diff: {log_prob_diff.item():.10f}")  # debug
                    log_prob_diff = torch.clamp(log_prob_diff, -20, 20)
                    ratio = torch.exp(log_prob_diff)

                    if torch.isnan(ratio) or torch.isinf(ratio):
                        print("  ratio is nan/inf, skipping update")
                        optimizer.zero_grad()
                        continue

                    r = sample["reward_normalized"]
                    loss_unclipped = ratio * r
                    loss_clipped = torch.clamp(ratio, 1 - DDPO_CFG["clip_range"],
                                                      1 + DDPO_CFG["clip_range"]) * r
                    policy_loss = -torch.min(loss_unclipped, loss_clipped)

                    kl = (ratio - 1) - torch.log(ratio + 1e-6)
                    loss = policy_loss + DDPO_CFG["kl_coeff"] * kl

                    loss.backward()

                    # add this temporarily
                    num_with_grad = sum(1 for p in pipe.transformer.parameters() if p.grad is not None and p.grad.abs().max() > 0)
                    print(f"  Params with nonzero grad: {num_with_grad}")

                    # torch.nn.utils.clip_grad_norm_(pipe.transformer.parameters(), DDPO_CFG["grad_clip"])
                    # optimizer.step()
                    w_before = next(p for p in pipe.transformer.parameters() if p.requires_grad).data.clone()
                    torch.nn.utils.clip_grad_norm_(pipe.transformer.parameters(), DDPO_CFG["grad_clip"])
                    optimizer.step()
                    w_after = next(p for p in pipe.transformer.parameters() if p.requires_grad).data.clone()
                    print(f"  Weight change: {(w_after - w_before).abs().max().item():.10f}")
                    if ((w_after - w_before).abs().max().item() < 1e-3):
                        print([n for n, p in pipe.transformer.named_parameters() if p.requires_grad][:5])

                    # log per-step metrics
                    wandb.log({
                        "epoch": epoch,
                        "inner_epoch": inner_epoch,
                        "loss": loss.item(),
                        "ratio": ratio.item(),
                        "log_prob_diff": log_prob_diff.item(),
                    })

                    total_loss += loss.item()
                    update_count += 1

                    print(f"  [inner {inner_epoch}] loss={loss.item():.4f}  ratio={ratio.item():.10f}")

                    del loss, ratio, kl, policy_loss, log_prob_new
                    torch.cuda.empty_cache()

                except torch.cuda.OutOfMemoryError:
                    print(f"  OOM during PPO update, skipping")
                    optimizer.zero_grad()
                    if 'prompt_embeds' in dir(): del prompt_embeds
                    if 'neg_embeds' in dir(): del neg_embeds
                    if 'log_prob_new' in dir(): del log_prob_new
                    if 'loss' in dir(): del loss
                    torch.cuda.empty_cache(); gc.collect()
                # add back later
                # except Exception as e:
                #     import traceback
                #     print(f"  Error during PPO update: {traceback.format_exc()}, skipping")
                #     optimizer.zero_grad()
                #     torch.cuda.empty_cache(); gc.collect()

        pipe.transformer.eval()

        if update_count > 0:
            wandb.log({
                "epoch": epoch,
                "avg_loss": total_loss / update_count,
            })
        del sample_buffer, video_paths
        gc.collect()
        torch.cuda.empty_cache()

    # ── Save checkpoint ───────────────────────────────────────────────────────
    if epoch % 1 == 0:
        try:
            pipe.transformer.save_pretrained(
                f"/content/drive/MyDrive/CS234_Project/checkpoints/wan_ddpo_2.5e-4_batch_16_epoch{epoch}"
            )
            print(f"  Saved checkpoint at epoch {epoch}")
        except Exception as e:
            print(f"  Failed to save checkpoint: {e}")

```
